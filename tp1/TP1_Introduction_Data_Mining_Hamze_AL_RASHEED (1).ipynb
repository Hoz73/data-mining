15. Créez un dataframe appelez *sydney* contenant uniquement les observations pour la ville de Sydney. **Tips:** vous devrez peut être reset les index.

```python
# Create a new dataframe with only Sydney
sydney = df[df['Location'] == "Sydney"]
pd.to_datetime(sydney['Date'], errors='coerce')
# And reset the index
sydney = sydney.reset_index()
sydney
```

```python
# MinTemp moyenne
print(sydney['MinTemp'].mean())

# MaxTemp median
print(sydney['MaxTemp'].median())

# Sum Rainfall
print(sydney['Rainfall'].sum())
```

18. Remplacer les valeurs manquantes pour Rainfall par la moyenne et pour MaxTemp par la médiane.

```python
# Replace NaN
# sydney['Rainfall'].isnull()
sydney['Rainfall'].fillna(value=sydney['MaxTemp'].mean(), inplace=True)
sydney['MaxTemp'].fillna(value=sydney['MaxTemp'].median(), inplace=True)
sydney.isnull().sum()
```

19. Créer une colonne RainToday qui prend comme valeur 1 si Rainfall>0 et 0 si Rainfall==0. **Tips:** vous pouvez faire une boucle.

```python
# Loop over sydney rows
tab = []
for row in sydney.itertuples():
  tab.append(1) if row.Rainfall > 0 else tab.append(0)


sydney['RainToday'] = tab

# Print sample of sydney
sydney.sample(5)
```

21. Créez la colonne *Pressure* représentant la valeur moyenne des valeurs de Pressure9am et Pressure3pm :  
Pressure[i] = (Pressure9am[i] + Pressure3pm[i])/2

```python
# Create Pressure column
sydney['Pressure'] = sydney[['Pressure9am', 'Pressure3pm']].mean(axis=1)
sydney
```

22. Tracer la variation de *Pressure* du mois de janvier 2016

```python
# Select dates
sydney['Date'] = sydney['Date'].astype('datetime64[ns]')
janvier2016 = sydney[(sydney['Date'].dt.year == 2016) & (sydney['Date'].dt.month == 1) ]
dates = janvier2016['Date']

# Select pressure
pressures = janvier2016['Pressure']

# # Plot size
plt.figure(figsize=(12,8))

# # Plot it
plt.plot(dates, pressures)

# # Plot customization
plt.xticks(dates, rotation='vertical')
plt.show()
```

24. Concaténez *sydney* et *canberra* suivant leurs lignes

25. [Standardisez](https://stackoverflow.com/questions/26414913/normalize-columns-of-pandas-data-frame) toutes les valeurs quantitatives de votre nouveau dataframe. **Tips:** plusieurs étapes sont nécessaires.

```python
# Standardization
temporary = concat_df[['Date', 'Location']]
concat_df = concat_df.drop(columns=['Date', 'Location'])

df_normalized = (concat_df - concat_df.min()) / (concat_df.max() - concat_df.min())
df_normalized['Location'] = temporary['Location']
df_normalized['Date'] = temporary['Date']

df_normalized.head(50)
```
